````python
import pandas as pd

# Configura√ß√£o de exibi√ß√£o
df = wc.copy()

# Convers√£o de tipos
df['data'] = pd.to_datetime(df['data'])  # Converter para datetime
df['ano'] = df['ano'].astype('int64')  # Garantir que ano seja int

# Convers√£o de colunas para string
df[['time_1', 'time_2', 'pa√≠s_sede', 'resultado', 'gols_1_detalhes', 'gols_2_detalhes', 
    'gols_1_contra', 'gols_2_contra', 'gols_1_penalti', 'gols_2_penalti',
    'cartao_vermelho_1', 'cartao_vermelho_2', 'cartao_amarelo_1', 'cartao_amarelo_2', 'copa']] = 
    df[['time_1', 'time_2', 'pa√≠s_sede', 'resultado', 'gols_1_detalhes', 'gols_2_detalhes', 
        'gols_1_contra', 'gols_2_contra', 'gols_1_penalti', 'gols_2_penalti',
        'cartao_vermelho_1', 'cartao_vermelho_2', 'cartao_amarelo_1', 'cartao_amarelo_2', 'copa']].astype('string')

# Converter colunas num√©ricas
num_cols = ['gols_1', 'gols_2', 'comparecimento']
df[num_cols] = df[num_cols].astype('int64')

# Converter 'rodada' para categoria
df['rodada'] = df['rodada'].astype('category')

# Encontrar jogo com maior audi√™ncia
jogo_maior_audiencia = df.loc[df['comparecimento'].idxmax()]
print(jogo_maior_audiencia)

# Contar n√∫mero de Copas Femininas e Masculinas
copas_contagem = df[['ano', 'copa']].drop_duplicates()['copa'].value_counts()
print(f"Masculina: {copas_contagem.get('Masculina', 0)}")
print(f"Feminina: {copas_contagem.get('Feminina', 0)}")

# Criar DataFrame de participa√ß√£o
participacao = (df[['time_1', 'copa', 'ano']]
                .drop_duplicates()
                .groupby(['time_1', 'copa'])
                .count()
                .reset_index()
                .rename(columns={'time_1': 'pa√≠s', 'ano': 'num_copas'}))

# Exibir top 5 pa√≠ses de cada competi√ß√£o
print(participacao[participacao['copa'] == 'Feminina'].nlargest(5, 'num_copas'))
print(participacao[participacao['copa'] == 'Masculina'].nlargest(5, 'num_copas'))

# Criar DataFrame de total de gols por pa√≠s
gols_1 = df.groupby('time_1')['gols_1'].sum().reset_index()
gols_2 = df.groupby('time_2')['gols_2'].sum().reset_index()
gols_1.columns = ['pa√≠s', 'total_gols']
gols_2.columns = ['pa√≠s', 'total_gols']
gols = pd.concat([gols_1, gols_2]).groupby('pa√≠s').sum().reset_index()
gols = gols.sort_values(by='total_gols', ascending=False)
print(gols)

# Fun√ß√£o para contar cart√µes amarelos
def contar_cartoes(cartoes):
    if pd.isna(cartoes):
        return 0
    return len(eval(cartoes))  # Avalia a string como lista e conta elementos

# Criar colunas de contagem de cart√µes amarelos
df['num_cartoes_1'] = df['cartao_amarelo_1'].apply(contar_cartoes)
df['num_cartoes_2'] = df['cartao_amarelo_2'].apply(contar_cartoes)

# Criar DataFrame de total de cart√µes amarelos por pa√≠s
cartoes_1 = df.groupby('time_1')['num_cartoes_1'].sum().reset_index()
cartoes_2 = df.groupby('time_2')['num_cartoes_2'].sum().reset_index()
cartoes_1.columns = ['pa√≠s', 'cartoes_amarelos']
cartoes_2.columns = ['pa√≠s', 'cartoes_amarelos']
cartoes = pd.concat([cartoes_1, cartoes_2]).groupby('pa√≠s').sum().reset_index()
cartoes = cartoes.sort_values(by='cartoes_amarelos', ascending=False)
print(cartoes)

# Fun√ß√£o para contar gols por jogador
def contar_gols(col):
    if pd.isna(col):
        return []
    return [gol.split(' ¬∑ ')[0] for gol in col.split('|')]

# Criar dicion√°rio para contagem de gols por jogador
gols_jogadores = {}
for col in ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']:
    df[col].dropna().apply(lambda x: [gols_jogadores.update({jogador: gols_jogadores.get(jogador, 0) + 1}) for jogador in contar_gols(x)])

# Criar DataFrame de gols por jogador
gols_jogadores_df = pd.DataFrame(gols_jogadores.items(), columns=['jogador(a)', 'num_gols'])
gols_jogadores_df = gols_jogadores_df.sort_values(by='num_gols', ascending=False)
print(gols_jogadores_df.head(10))

# Salvar DataFrame formatado
df.to_csv('wc_formatado.csv', index=False)

# Exibir informa√ß√µes do DataFrame formatado
df.info()
````
